# Fungsi yang dipakai

```python
# mencari rata-rata rating tiap user
def mean_rating(rating_table):
    mean = np.mean(rating_table, axis=1)
    
    return mean
```

```python
# normalisasi rating user
def normalized_rating(mean):
    normalized = rating_table.subtract(mean, axis=0)
    
    return normalized
```

```python
# mengisi value NaN agar dapat dihitung cosine similaritynya
def fillna_rating_normalized(normalized):
    normalized_fillna = normalized.fillna(0)
    
    return normalized_fillna
```

```python
# menghitung cosine similarity data user yang ingin direkomendasikan dengan user lainnya
def cosine_similarity_user(normalized_fillna, user):
    cs = cosine_similarity(normalized_fillna, [normalized_fillna.loc[user]]) #user_id
    
    return cs
```

```python
# mengurutkan index cosine similarity
def sort_cosine_similarity_user(cs):
    nearest_index = np.argsort(-cs, axis=0)

    # ambil 50 user id dengan similarity terdekat
    # index 0 tidak terpakai karena pasti merujuk ke data user ID yang ingin diprediksi
    nearest_index = nearest_index[1:51]
    
    return nearest_index
```

```python
# mengambil mean data user yang ingin diprediksi
def find_mean_user(mean, user):
    mean_user = mean.loc[user] #user_id
    
    return mean_user
```

```python
# mengambil cosine similarity dari 50 user terdekat similaritynya
def nearest_cosine_similarity_user(cs, nearest_index):

    nearest_cs = []

    for i in range (50):
        nearest_cs.append(cs[nearest_index[i]])

    nearest_cs = np.array(nearest_cs).reshape(-1,1)
    
    return nearest_cs
```

```python
# mencari id buku yang masih belum dirating oleh user yang ingin direkomendasikan
def find_isna_rating(rating_table, user, isna=True):
    rating_isna = rating_table.loc[user][rating_table.loc[user].isna()==isna].index #user_id
    
    return rating_isna
```

```python
# mengambil data rating buku yang ingin diprediksi dari user terdekat
def find_book_nearest_user(rating_isna, normalized_fillna, nearest_index):
    nearest_rating = []
    all_nearest_rating = []

    for i in range(50):
    #     print(i+1, end=" ")
        for j in rating_isna:
            nearest_rating.extend(normalized_fillna.loc[nearest_index[i]+1][j]) # index nearest mulai dari 0 sehingga harus ditambah 1 agar sesaui id user

        all_nearest_rating.append(nearest_rating)
        nearest_rating = []

    all_nearest_rating = np.array(all_nearest_rating)
    
    return all_nearest_rating
```

```python
# menghitung prediksi rating dengan rumus ppt
def predict_rating(rating_isna, mean_user, nearest_cs, all_nearest_rating):
    prediction = []

    for i in range (len(rating_isna)):
        # sumber:
        # ppt sesi 7
        # https://github.com/tttgm/fellowshipai/blob/master/Book-Crossing-Recommender.ipynb
        # prediction_temp = mean_user_rating + (sum(cosine_similarity * normalized_rating) / sum(cosine_similarity))
        prediction_temp = mean_user+(nearest_cs.T.dot(all_nearest_rating[:,i]) / np.sum(nearest_cs))
        prediction.append(prediction_temp)

    return prediction
```

```python
# mengurutkan prediksi
def sort_rating(prediction):
    prediction_index = np.argsort(-np.ravel(prediction))
    
    return prediction_index
```

```python
def print_recommended(rating_isna, prediction_index, book_data, prediction):
    print("Highly Recommended")
    for i in range (5):
        rec_book_id = rating_isna[prediction_index[i]]
        print(f"{i+1}. {rec_book_id} - {book_data.title[rec_book_id]}", end=" ") 
        print(f"by {book_data.author[rec_book_id]} - {book_data.year[rec_book_id]}")
        print(f"   (rating prediction = {prediction[prediction_index[i]][0]})\n")
    
```

# Prediksi dan beri rekomendasi dari rating tertinggi

```python
def collaborative_filtering_recommendation(rating_table, book_data, user, isna=True, isprint=True):
    print(f"user {user}")
    print("Please Wait", end="")
    mean = mean_rating(rating_table)
    print("-", end="")
    normalized = normalized_rating(mean)
    print("-", end="")
    normalized_fillna = fillna_rating_normalized(normalized)
    print("-", end="")
    cs = cosine_similarity_user(normalized_fillna, user)
    
    print("-", end="")
    nearest_index = sort_cosine_similarity_user(cs)
    print("-", end="")
    nearest_cs = nearest_cosine_similarity_user(cs, nearest_index)
    
    print("-", end="")
    mean_user = find_mean_user(mean, user)
    
    print("-", end="")
    rating_isna = find_isna_rating(normalized,user,isna)
    print("-", end="")
    all_nearest_rating = find_book_nearest_user(rating_isna, normalized_fillna, nearest_index)
    
    print("-", end="")
    prediction = predict_rating(rating_isna, mean_user, nearest_cs, all_nearest_rating)
    
    print("-\n")
    if(isprint):
        prediction_index = sort_rating(prediction)
        print_recommended(rating_isna, prediction_index, book_data, prediction)
    
    return prediction
```

```python
result = collaborative_filtering_recommendation(rating_table, book_data, user)
```
